```python
list = []
```

```python
for i in range(10):
  angka = int(input("masukan angka: "))
  list.append(angka)

```

```python
print("list: ", list)
print("nilai maksimum:", max(list))
print("nilai minimum:", min(list))
print("rata-rata:", sum(list)/len(list))
```

```python
import matplotlib.pyplot as plt
plt.bar(range(len(list)), list)
plt.title("grafik nilai")
plt.xlabel("indeks")
plt.ylabel("nilai")
plt.show()
```
